### Explicação questão 1

    I. A diferença entre árvore binária de busca e árvores AVL é o fato de que a
       segunda pode se reconfigurar dinamicamente, com o intuito de manter um
       bom nível de balanceamento.

A **ABB (árvore de busca binária)** não possui nenhuma forma de balanceamento, sendo esse um de seus problemas de implementação para uma grande quantidade de dados. A **árvore AVL (Adelson Velsky e Landis)** tem como propósito balancear a ABB, ou seja, ela é reconfigurada dinâmicamente quando novos itens são inseridos. Isso para manter a árvore balanceada.

A diferença entre a ABB e AVL é que a **AVL possui balanceamento e a ABB não**

------

    II. Uma pilha garante que o último elemento inserido seja localizado no seu
        topo. Porém, do ponto de vista conceitual, qualquer elemento da pilha pode
        ser removido, ainda que não esteja no seu topo.

De fato a pilha segue o padrão *Last In First Out (LIFO)*. Seguindo o princípio LIFO (que é conceitual) **somente o topo pode ser removido** (último a entrar primeiro a sair).

[pilha](https://upload.wikimedia.org/wikipedia/commons/b/b4/Lifo_stack.png "Pilha")

----

    III. Do ponto de vista conceitual, não há diferença alguma entre uma estrutura
         de array e uma lista encadeada.

O array conceitualmente está organizado de forma contínua na memória, ou seja, não há espaços de memória entre os itens de um array. Por outro lado, uma lista encadeada pode está distribuida de forma aleatória na memória. 

Dessa forma, arrays e lista encadeadas **não são iguais do ponto de vista conceitual.**

---

    IV. Tabelas hash são estruturas de dados indicadas para armazenar grande
        volume de dados. Apesar dessas estruturas permitirem acesso indexado,
        mais de um elemento pode ter o mesmo índice. Elementos com o mesmo
        índice podem ser armazenados em uma mesma lista encadeada.

De fato tabelas hash são indicados para armazenar grande quantidade de dados, devido sua complexidade ser contante. No caso de elementos com o mesmo índice (colisão), **pode-se armazenar esses itens em uma lista encadeada**.
